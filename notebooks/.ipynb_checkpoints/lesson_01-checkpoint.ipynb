```python
import pandas as pd

from datetime import datetime
```

```python
df = pd.read_csv('../data/lesson_1_data.csv', encoding='utf-8', sep=',')
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.rename(columns={
    'номер заказа': 'order_id',
    'создан': 'created_at',
    'оплачен': 'paid_at',
    'название курса': 'course_name',
    'статус': 'status',
    'заработано': 'revenue_rub',
    'город': 'city',
    'платежная система': 'payment_system'
})
```

```python
df.columns
```

```python
df.course_name.unique()
```

```python
all_revenue_rub = df['revenue_rub'].sum()
```

```python
all_revenue_rub
```

```python
df.groupby('course_name', as_index=False) \
    .aggregate({'revenue_rub': 'sum'})
```

```python
revenue_rub_by_city_and_courses = df.groupby(['course_name', 'city']) \
    .aggregate({'revenue_rub': 'sum'}) \
    .sort_values('revenue_rub', ascending=False)
```

```python
revenue_rub_by_city_and_courses.to_csv('revenue_rub_by_city_and_courses.csv')
```

```python
money_by_title = df \
    .query("status == 'завершен'") \
    .groupby('course_name', as_index=False) \
    .aggregate({
        'revenue_rub': 'sum',
        'order_id': 'count',
    }) \
    .sort_values('revenue_rub', ascending=False) \
    .rename(columns={'order_id': 'success_orders'})

```

```python
today_day = datetime.today().strftime('%Y-%m-%d')
filename = f'money_title_{today_day}.csv'
```

```python
if int(money_by_title.revenue_rub.sum()) == int(all_revenue_rub):
    print(f'OK. File {filename} is written.')
else:
    print('ERROR!')
```
